```python
acura = engine.execute("SELECT car_year, make, model, subseries, color, condition_grade, mileage FROM acura")

cars = acura.fetchall()

car_year = []
make = []
model = []
subseries = []
color = []
condition_grade = []
mileage = []

for row in cars:
    car_year.append(row[0])
    make.append(row[1])
    model.append(row[2])
    subseries.append(row[3])
    color.append(row[4])
    condition_grade.append(row[5])
    mileage.append(row[6])

cars_dict = {'car_year':car_year,
            'make':make,
            'model':model,
            'subseries':subseries,
            'color':color,
            'condition_grade':condition_grade,
            'mileage':mileage}
cars_dict
```

```python
%%time
def unique_val(col_name):
        df = pd.DataFrame()
        for table in tables:
            data = pd.read_sql_query(f"SELECT DISTINCT {col_name} FROM {table}", con = engine)
            df = pd.concat([df, data], ignore_index = True)
        res = df.sort_values(by = df.columns[0], ascending = False)[df.columns[0]].unique()
        return res

carData = {'car_year': unique_val('car_year'),
            'make':unique_val('make'),
            'model':unique_val('model'),
            'subseries':unique_val('subseries'),
            'color':unique_val('color'),
            'condition_grade':unique_val('condition_grade'),
            'mileage':unique_val('mileage')}
carData
```
